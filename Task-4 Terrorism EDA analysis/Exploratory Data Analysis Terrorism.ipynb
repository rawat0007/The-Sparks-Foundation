**Data Dictionary**

* 'iyear' : year of the incident

* 'imonth' : month of the incident

* 'iday' : day of the incident

* 'nkill' : number of dead

* 'nwound' : number of wounded

* extended' : has the incident lasted for more than 24 hours?

* 'multiple' : is the incident connected to other attacks?

* 'success' : did the terrorist attack achieve its goal (i.e. assassination, etc.)?

* 'suicide' : did the incident involve a suicide attack?

* 'guncertain1' : was the terrorist group confirmed?

* 'attacktype1_txt' : general method of attack used (i.e. assassination, hijacking, bombing/explosion, etc.)

* targtype1_txt' : general type of target/victim (i.e. business, government, police, military, etc.)

* 'natlty1_txt' : nationality of the target/victim

* 'weaptype1_txt' : general type of weapon used in the incident (i.e. biological, chemical, firearms, etc.)

* 'target1' : specific person, building, installation, etc. that was targeted "

# REPLACING THE NULL VALUES

```python
data["state"].replace(np.nan,0,inplace=True)
data["city"].replace(np.nan,0,inplace=True)
data["latitude"].replace(np.nan,0,inplace=True)
data["longitude"].replace(np.nan,0,inplace=True)
data["summary"].replace(np.nan,0,inplace=True)
data["targetsubtype"].replace(np.nan,0,inplace=True)
data["motive"].replace(np.nan,0,inplace=True)
data["weapsubtype"].replace(np.nan,0,inplace=True)
data["kill"].replace(np.nan,0,inplace=True)
data["wound"].replace(np.nan,0,inplace=True)
data["propertyvalue"].replace(np.nan,0,inplace=True)
data["propcomment"].replace(np.nan,0,inplace=True)
```

# Country with highest terrorist attack

```python
abz= plt.subplots(figsize=(16,11))
sns.barplot(x=data["country"].value_counts()[:20].values,y=data["country"].value_counts()[:20].index,palette="cool")
plt.title("Country With Highest Terrorist Attack")
plt.show()
```

```python
# from above graph we can conclude that IRAQ has highest terrorist attack
```

# State With Highest Terrorist Attack

```python
sns.countplot(data=data,x="state",order=data["state"].value_counts().iloc[:20].index)
plt.title("State With Highest Terrorist Attack")
plt.xticks(rotation=90)
plt.show()
```

```python
# from above graph we can conclude that STATE BAGHDAD has highest terrorist attack
```

# Highest Targeted Region

```python
xyz= data["region"].value_counts().iloc[:10]
xyz.plot(kind="pie",autopct="%0.2f%%",radius=2.0,)
plt.show()
```

```python
# from above piechart we can conclude that REGION MIDDLE EAST AND NORTH AFRICA has highest terrorist attack
```

# Most Used Weapon Type

```python
plt.title("WEAPONS USED")
sns.countplot(data=data,x="weapontype",order=data["weapontype"].value_counts().iloc[:5].index)
plt.show()
```

```python
# EXPLOSIVES are highly used weapons in the terrorist attack
```

# ATTACK TYPE

```python
plt.title("Method Of Attack")
sns.histplot(x=data["attacktype"],data=data,kde=True,color="red")
plt.xticks(rotation=90)
plt.show()
```

```python
# Most of the attacks are through BOMBING AND EXPLOSION
```

# Highly Targeted Places

```python
plt.title("Highly Targeted Places")
sns.histplot(x=data["targetype"],data=data,kde=True,color="green")
plt.xticks(rotation=90)
plt.show()
```

```python
# Private citizens and property > Military > Police > Government
```

# TERRORIST ORGANISATIONS

```python
xyz= data["gangname"].value_counts().iloc[1:10]
myexplode=[0.2,0,0,0,0,0,0,0,0]
xyz.plot(kind="pie",autopct="%0.2f%%",radius=2.0,explode=myexplode)
plt.show()
```

```python
# Taliban is the active terrorist organisation in the terrorist attack
```

# REGION WITH HIGHEST DEATHS

```python
sns.barplot(x=data["region"].value_counts()[:10].index,y=data["kill"].value_counts()[:10].index)
plt.xticks(rotation=90)
plt.show()
```
